```python
# ================================
# IMPORT REQUIRED LIBRARIES
# ================================
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# =========================================
# QUESTION 1: Line Chart of Alphabet Inc Data
# =========================================

# Creating dataframe manually (same as fdata.csv)
data = {
    'Date': ['10-03-16','10-04-16','10-05-16','10-06-16','10-07-16'],
    'Open': [774.25,776.03,779.30,779.00,779.65],
    'High': [776.06,778.71,782.07,780.47,779.65],
    'Low': [769.50,772.89,775.65,775.53,770.75],
    'Close': [772.56,776.43,776.46,776.85,775.08]
}

df = pd.DataFrame(data)

# Plot
plt.figure()
plt.plot(df['Date'], df['Open'], label='Open')
plt.plot(df['Date'], df['High'], label='High')
plt.plot(df['Date'], df['Low'], label='Low')
plt.plot(df['Date'], df['Close'], label='Close')

plt.title("Alphabet Inc Stock Data")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()


# =========================================
# QUESTION 2: Horizontal Bar Plot
# =========================================

nos = [2, 9, 20, 25, 30, 39]
marks = [12, 24, 25, 27, 29, 30]

plt.figure()
plt.barh(nos, marks)
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.title("Students vs Marks")
plt.show()


# =========================================
# QUESTION 3: Year-wise Sales Line Chart
# (Based on table from PDF page 1)
# =========================================

months = ['Jan','Feb','Mar','Apr','May','Jun']

sales_2018 = [2500,2200,3000,2800,2750,3500]
sales_2019 = [2600,1800,2500,3000,3200,2000]
sales_2020 = [2450,2000,3000,3100,1950,3550]
sales_2021 = [3250,3200,3800,2800,2900,2300]

plt.figure()
plt.plot(months, sales_2018, label='2018')
plt.plot(months, sales_2019, label='2019')
plt.plot(months, sales_2020, label='2020')
plt.plot(months, sales_2021, label='2021')

plt.title("Year-Wise Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()
plt.show()


# =========================================
# QUESTION 4: Customized Line Chart
# =========================================

plt.figure()
plt.plot(months, sales_2021, marker='*', markersize=10, linestyle='--', linewidth=3)

plt.title("Customized Sales Chart")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()


# =========================================
# QUESTION 5: Bar Chart (Employee Data)
# =========================================

dept = ['HR','Sales','Finance','Marketing','IT']
emp = [40,30,35,32,45]

plt.figure()
plt.bar(dept, emp, color='red')

plt.title("Compu-Tech Employee Data")
plt.xlabel("Departments")
plt.ylabel("Number of Employees")
plt.show()


# =========================================
# QUESTION 6: Scatter + Grouped Bar Chart
# =========================================

# Sample data (since real dataset not provided)
data = {
    'State': ['A','B','C','D','E','F'],
    'Population': [10,20,30,40,50,60],
    'Literacy': [70,75,80,85,65,90],
    'Region': ['North','South','East','West','North','South']
}

df = pd.DataFrame(data)

# Scatter plot by region
plt.figure()
for region in df['Region'].unique():
    temp = df[df['Region'] == region]
    plt.scatter(temp['Population'], temp['Literacy'],
                s=np.sqrt(temp['Literacy'])*10,
                marker='D', label=region)

plt.xlabel("Population")
plt.ylabel("Literacy Rate")
plt.title("Population vs Literacy")
plt.legend()
plt.show()

# Average literacy bar chart
avg = df.groupby('Region')['Literacy'].mean()

plt.figure()
avg.plot(kind='bar')
plt.title("Average Literacy Rate by Region")
plt.show()


# =========================================
# QUESTION 7: Box Plot (Student Marks)
# =========================================

marks_data = {
    'English':[95,95,78,88],
    'Maths':[95,76,81,63],
    'Hindi':[90,79,75,67],
    'Science':[94,77,76,77],
    'Social':[95,89,88,80]
}

df = pd.DataFrame(marks_data)

plt.figure()
df.boxplot()
plt.title("Student Performance Comparison")
plt.show()


# =========================================
# QUESTION 8: Bar Plot (Men vs Women)
# =========================================

men = [22,30,35,35,26]
women = [25,32,30,35,29]

x = np.arange(len(men))

plt.figure()
plt.bar(x-0.2, men, width=0.4, label='Men')
plt.bar(x+0.2, women, width=0.4, label='Women')

plt.xlabel("Groups")
plt.ylabel("Scores")
plt.title("Scores by Gender")
plt.legend()
plt.show()


# =========================================
# QUESTION 9: Pie Chart (Sales Data)
# =========================================

products = ['A','B','C','D']
sales = [50,30,10,10]

plt.figure()
plt.pie(sales, labels=products, autopct='%1.1f%%')
plt.title("Sales Distribution")
plt.show()


# =========================================
# QUESTION 9(b): Subplots (Bathing soap & Facewash)
# =========================================

months = list(range(1,13))
soap = [9000,6000,9500,8500,7500,7400,9000,10000,8000,10500,13000,14000]
facewash = [1500,1200,1300,1100,1750,1550,1100,1400,1800,1900,2100,1750]

plt.figure()

plt.subplot(2,1,1)
plt.plot(months, soap)
plt.title("Sales of Bathing Soap")

plt.subplot(2,1,2)
plt.plot(months, facewash, color='red')
plt.title("Sales of Facewash")

plt.show()


# =========================================
# QUESTION 10: Histogram (Profit Data)
# =========================================

profit = [200000,250000,180000,300000,220000,270000,210000,350000]

plt.figure()
plt.hist(profit)
plt.title("Profit Data Histogram")
plt.xlabel("Profit Range")
plt.ylabel("Frequency")
plt.show()
```
